```python
 import numpy as np
 import pandas as pd
 import difflib
 from sklearn.feature_extraction.text import TfidfVectorizer
 from sklearn.metrics.pairwise import cosine_similarity
```

```python
# importing dependents
# reading data
# data shape
# selecting features
# vectorizer
# similarity
# input data
# # indexing
```

```python
data=pd.read_csv("movies.csv")
```

```python
data.head()
```

```python
data.shape
```

```python

selected_features = ['genres','keywords','tagline','cast','crew','director']
print(selected_features)
```

```python
for features in selected_features:
  data[features] = data[features].fillna('')
```

```python
combined_features = data['genres']+' '+data['keywords']+' '+data['tagline']+' '+data['cast']+' '+data['crew']+' '+data['director']
print(combined_features)
```

```python
feature_vector = TfidfVectorizer()
```

```python
vector = feature_vector.fit_transform(combined_features)
```

```python
print(vector)
```

```python
similarity = cosine_similarity(vector)
```

```python
print(similarity)
```

```python
similarity.shape
```

```python
movie_input = input('Enter the movie : \n')
```

```python
list_of_movies = data['title'].tolist()
```

```python
print(list_of_movies)
```

```python
find_close_match = difflib.get_close_matches(movie_input,list_of_movies)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

```python
index_of_the_movie = data[data.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

```python
similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
sorted_data = sorted(similarity_score,key = lambda x:x[1],reverse = True)
print(sorted_data)
```

```python
print("The recommended movies for you: \n")

i = 1

for movie in sorted_data:
  index = movie[0]
  title_of_the_movie = data[data.index==index]['title'].values[0]
  if (i<31):
    print(i,'.',title_of_the_movie)
    i+=1
```
